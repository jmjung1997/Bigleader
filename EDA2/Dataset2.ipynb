###  입지계수: A지역 특정 산업군 종사자 수/ A지역 전체 종사자 수     /  전 지역군 특정 산업군 종사자 수/전체 종사자수

```python
jongsa= pd.read_csv('./data/co_p.csv',encoding='utf-8')

jongsa.drop(['Unnamed: 0'], axis = 1, inplace = True)
jongsa.shape



jongsa_g = jongsa.groupby(['STDR_YM','CTPRVN_NM','SIGNGU_NM','산업_중분류'], as_index=False).sum()

jongsa_g.head()

jongsa_g_san = jongsa.groupby(['STDR_YM','산업_중분류'], as_index=False).sum()


city_list=jongsa_g['SIGNGU_NM'].values.tolist()
sanup_list=jongsa_g['산업_중분류'].values.tolist()
date_list=jongsa_g['STDR_YM'].values.tolist()

ypji=[]
for i, j, z in zip(city_list, sanup_list, date_list):
        city=i
        sanup=j
        date=z
        jongsa_sum=jongsa_g[jongsa['STDR_YM']==date]['피보험자수'].sum() #각년 전체 종사자 수
        sum_region=jongsa_g[(jongsa['STDR_YM']==date)&(jongsa['산업_중분류']==sanup)]['피보험자수'].sum() #각 년도에 전 지역군 산업별 종사자수
        region_co=jongsa_g[(jongsa['STDR_YM']==date)&(jongsa['SIGNGU_NM']==city)& (jongsa['산업_중분류']==sanup)]['피보험자수'].sum() #각 년도 특정 지역 특정 산업군 종사자 수
        region_sum=jongsa_g[(jongsa['STDR_YM']==date)&(jongsa['SIGNGU_NM']==city)]['피보험자수'].sum() #각 년도 특정 지역 종사자 수
        
        result=(region_co/region_sum)/(sum_region/jongsa_sum)
        ypji.append(result.round(4))
```

```python
ypji_1=ypji
```

```python
data = {
    '입지계수':ypji
}
```

### 대분류 입지계수 구하기

```python
appliance= pd.read_excel('./data/mismatch/ws_insuarance.xlsx')
```

```python
appliance['마감년월'].fillna(method='ffill', inplace=True)
appliance['시군구'].fillna(method='ffill', inplace=True)
appliance.head()
```

```python
appliance['산업_대분류']=appliance['산업_대분류'].str.replace('10차_','')
```

```python
appliance.head()
```

```python
appliance['마감년월']=appliance['마감년월'].str.replace('년','-').str.replace('월','')
```

```python
split = appliance.시군구.str.split(' ')
split = split.apply(lambda x: pd.Series(x))
split

tmp = pd.concat([appliance, split], axis=1)
```

```python
tmp.head()
```

```python
colnames = list(tmp.columns)

colnames[4:] = ['시도명','시군구명']
tmp.columns = colnames
tmp.fillna('세종시', inplace=True)
```

```python
tmp.columns
```

```python
appliance_new = tmp[['마감년월','시도명', '시군구명', '산업_대분류', '피보험자수']]
appliance_new.head()
```

```python
appliance_new['마감년월']=pd.to_datetime(appliance_new['마감년월'], format='%Y-%m')
```

```python
appliance_new.columns = ['STDR_YM','CTPRVN_NM','SIGNGU_NM','산업_대분류','피보험자수']
```

```python
appliance_new.head()
```

```python
appliance_new.shape
```

```python
appliance_g.shape
```

```python
appliance_g = appliance_new.groupby(['STDR_YM','CTPRVN_NM','SIGNGU_NM','산업_대분류'], as_index=False).sum()
appliance_g.columns = ['STDR_YM','CTPRVN_NM','SIGNGU_NM','산업_대분류','A_지역_특정_산업']
appliance_g.head()
```

```python
appliance_g_san = appliance_new.groupby(['STDR_YM','산업_대분류'], as_index=False).sum()
appliance_g_san.columns = ['STDR_YM','산업_대분류','전_지역_특정_산업']
appliance_g_san.head()
```

```python
appliance_g_jong = appliance_new.groupby(['STDR_YM', 'CTPRVN_NM','SIGNGU_NM'], as_index=False).sum()
appliance_g_jong.columns = ['STDR_YM','CTPRVN_NM','SIGNGU_NM','A_지역_전체']
appliance_g_jong
```

```python
appliance_g_all = appliance_new.groupby(['STDR_YM'], as_index=False).sum()
appliance_g_all.columns = ['STDR_YM','전체']
appliance_g_all
```

```python
merged1 = pd.merge(appliance_g, appliance_g_all, how='left', on='STDR_YM')

merged2 = pd.merge(merged1, appliance_g_jong, how='left', on=['STDR_YM','CTPRVN_NM','SIGNGU_NM'])

merged_fin = pd.merge(merged2, appliance_g_san, how='left', on=['STDR_YM','산업_대분류'])

merged_fin['입지계수'] = (merged_fin['A_지역_특정_산업'] / merged_fin['A_지역_전체']) / (merged_fin['전_지역_특정_산업'] / merged_fin['전체'])

merged_fin.to_csv('./appliance_p.csv', encoding='utf-8', index=False)
```
