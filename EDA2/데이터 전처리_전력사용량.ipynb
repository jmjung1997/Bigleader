```python
import numpy as np
import pandas as pd 

data_elec=pd.read_excel('./data/extra/junggi.xlsx' )

data_elec


idx_nm_1=data_elec[data_elec['업종별']=='합 계'].index
el=data_elec.drop(idx_nm_1)
    


idx_nm_1

el['elec_use']=np.NaN

el.reset_index(inplace=True, drop=True)

el.head(40)

el_new=el
list_data=[]
list_data2=[]
list_data3=[]
list_data4=[]
list_data5=[]
for i in range(len(el_new)):
    for k in range(len(el.iloc[:,4:16].columns)):
        list_data.append(el.loc[i,str(k+1)+'월'])
        list_data2.append(str(el.loc[i,'연도'])+'-'+str(k+1))
        list_data3.append(el.loc[i,'시도'])
        list_data4.append(el.loc[i,'시군구'])
        list_data5.append(el.loc[i,'업종별'])
        
#     el.iloc[:,4:]

 list_data4

len(list_data)

data_df = pd.DataFrame(list_data)
data_df2= pd.DataFrame(list_data2)
data_df3= pd.DataFrame(list_data3)
data_df4= pd.DataFrame(list_data4)
data_df5= pd.DataFrame(list_data5)


df_list = [data_df,data_df2,data_df3,data_df4,data_df5] #데이터 프레임 합치기
df_all = pd.concat(df_list, ignore_index=True,axis=1)
df_all[0]

df = df_all[[1, 2, 3, 4,0]]

df.head()

df.columns = ['STDR_YM','CTPRVN_NM','SIGNGU_NM','INDUTY_MLSFC_NM','elec_use']

df.head()

df.to_excel('elec_use.xlsx',index=False)
```
