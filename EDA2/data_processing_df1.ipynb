```python
def to_signgu(df, signgu_col_name):
    date = df.마감년월.str.replace('년','-').str.replace('월','')
    split = df[signgu_col_name].str.split(' ')
    split = split.apply(lambda x: pd.Series(x))
    
    df.마감년월 = date
    
    tmp = pd.concat([df, split], axis=1)
    colnames = list(tmp.columns)
    colnames[-3:] = ['CTPRVN_NM','SIGNGU_NM', '일반구']
    colnames[0] = 'STDR_YM'
    tmp.columns = colnames
    tmp.drop('일반구', axis=1, inplace=True)
    tmp.fillna('세종시', inplace=True)
    df = tmp.groupby(['STDR_YM','CTPRVN_NM','SIGNGU_NM'], as_index=False).sum()
    df.drop(signgu_col_name, axis=1, inplace=True)
    return df
```

```python
sd_name = ['강원도','경기도','경상남도','경상북도','광주광역시','대구광역시',
         '대전광역시','부산광역시','서울특별시','세종특별자치시','울산광역시',
         '인천광역시','전라남도','전라북도','제주도','충청남도','충청북도']
f_name = []

for name in sd_name:
    f_name.append('./data/수출입실적/수출입 실적_' + name +'.xlsx')

for i in range(len(f_name)):
    if i == 0:
        df = pd.read_excel(f_name[i])
        df['시도명'] = sd_name[i]
        continue
    tmp = pd.read_excel(f_name[i])
    tmp['시도명'] = sd_name[i]
    df = pd.concat([df, tmp])
```
