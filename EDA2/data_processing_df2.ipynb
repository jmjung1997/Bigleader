### 기업 현황 그룹화

```python
grouped_entr = entr_info.groupby(['STDR_YM', 'CTPRVN_NM','SIGNGU_NM','INDUTY_LCLAS_CODE','INDUTY_MLSFC_CODE','INDUTY_LCLAS_NM','INDUTY_MLSFC_NM'], as_index=False).sum()
```

```python
grouped_entr
```

```python
grouped_entr['중견기업수'] = grouped_entr.MIDPZ_CO + grouped_entr.PRTC_TRGET_MIDPZ_CO
```

```python
grouped_entr['중소기업수'] = grouped_entr.MSLENTR_CO + grouped_entr.TLSME_CO + grouped_entr.SSLENTR_CO + grouped_entr.SCHTWR_CO + grouped_entr.DCS_EXCL_ENTRPRS_CO + grouped_entr.UN_CL_ENTRPRS_CO
```

```python
grouped_entr.head()
```

```python
use_cols = ['STDR_YM','CTPRVN_NM','SIGNGU_NM','INDUTY_LCLAS_CODE','INDUTY_MLSFC_CODE','INDUTY_LCLAS_NM','INDUTY_MLSFC_NM','LTRS_CO','중견기업수','중소기업수']
```

```python
df = grouped_entr[use_cols]
```

```python
df.head()
```

```python
entr_processed = df[df['STDR_YM'] >= '2015-11']
```

```python
entr_processed.reset_index(drop=True, inplace=True)
```

```python
grouped_emply = emply.groupby(['STDR_YM', 'CTPRVN_NM','SIGNGU_NM','INDUTY_LCLAS_CODE','INDUTY_MLSFC_CODE','INDUTY_LCLAS_NM','INDUTY_MLSFC_NM'], as_index=False).sum()
```

### 두 데이터셋 병합

```python
entr_processed.reset_index(drop=True, inplace=True)
emply_processed.reset_index(drop=True, inplace=True)
```

```python
merged_outer = pd.merge(entr_processed, emply_processed, how='outer',on=['STDR_YM','CTPRVN_NM','SIGNGU_NM','INDUTY_LCLAS_CODE','INDUTY_MLSFC_CODE','INDUTY_LCLAS_NM','INDUTY_MLSFC_NM'])
merged_inner = pd.merge(entr_processed, emply_processed, on=['STDR_YM','CTPRVN_NM','SIGNGU_NM','INDUTY_LCLAS_CODE','INDUTY_MLSFC_CODE','INDUTY_LCLAS_NM','INDUTY_MLSFC_NM'])
```

```python
merged_inner
```

```python
merged_inner.to_csv('./data/병합데이터셋(결측값X).csv', encoding='cp949', index=False)
merged_outer.to_csv('./data/병합데이터셋(결측값O).csv', encoding='cp949', index=False)
```
