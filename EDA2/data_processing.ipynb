### 전기 사용량 전처리

```python
import numpy as np
import pandas as pd 

data_elec=pd.read_excel('./data/extra/junggi.xlsx' )

data_elec


idx_nm_1=data_elec[data_elec['업종별']=='합 계'].index
el=data_elec.drop(idx_nm_1)
    


idx_nm_1

el['elec_use']=np.NaN

el.reset_index(inplace=True, drop=True)

el.head(40)

el_new=el
list_data=[]
list_data2=[]
list_data3=[]
list_data4=[]
list_data5=[]
for i in range(len(el_new)):
    for k in range(len(el.iloc[:,4:16].columns)):
        list_data.append(el.loc[i,str(k+1)+'월'])
        list_data2.append(str(el.loc[i,'연도'])+'-'+str(k+1))
        list_data3.append(el.loc[i,'시도'])
        list_data4.append(el.loc[i,'시군구'])
        list_data5.append(el.loc[i,'업종별'])
        
#     el.iloc[:,4:]

 list_data4

len(list_data)

data_df = pd.DataFrame(list_data)
data_df2= pd.DataFrame(list_data2)
data_df3= pd.DataFrame(list_data3)
data_df4= pd.DataFrame(list_data4)
data_df5= pd.DataFrame(list_data5)


df_list = [data_df,data_df2,data_df3,data_df4,data_df5] #데이터 프레임 합치기
df_all = pd.concat(df_list, ignore_index=True,axis=1)
df_all[0]

df = df_all[[1, 2, 3, 4,0]]

df.head()

df.columns = ['STDR_YM','CTPRVN_NM','SIGNGU_NM','INDUTY_MLSFC_NM','elec_use']

df.head()

df.to_excel('elec_use.xlsx',index=False)
```

## 여러 전처리 된 변수 데이터를 하나로 합치기

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

insurance= pd.read_excel('./data/extra/insurance_final3.xlsx')
silup= pd.read_excel('./data/extra/silup_clean_v2.xlsx')
swj = pd.read_excel('./data/extra/swj_birthday.xlsx')
elec= pd.read_excel('./data/extra/Use_elec.xlsx')


sum_join= pd.read_csv('./data/total/sum_join.csv' ,encoding='cp949')

sum_join.head()

sum_join.isnull().sum()

sum_join.reset_index(drop=True, inplace=True)
insurance.reset_index(drop=True, inplace=True)
silup.reset_index(drop=True, inplace=True)
swj.reset_index(drop=True, inplace=True)
elec.reset_index(drop=True, inplace=True)

merged_outer = pd.merge(sum_join, insurance, how='left',on=['STDR_YM','CTPRVN_NM','SIGNGU_NM'])
merged_inner1 = pd.merge(merged_outer, silup,how='left', on=['STDR_YM','CTPRVN_NM','SIGNGU_NM'])
merged_inner2 = pd.merge(merged_inner1, swj,how='left', on=['STDR_YM','CTPRVN_NM','SIGNGU_NM'])
merged_inner3 = pd.merge(merged_inner2, elec,how='left',on=['STDR_YM','CTPRVN_NM','SIGNGU_NM'])

merged_inner3

merged_inner3[merged_inner3['수급자격신청자수'].isnull()]

one_data=merged_inner3

one_data[one_data['수급자격신청자수'].isnull()]

one_data.to_excel('one_data.xlsx',index=False) #하나의 데이터셋으로 취합 완료 

elec[elec['SIGNGU_NM']=='미추홀구']
```
